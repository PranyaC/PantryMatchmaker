```python
df1 = pd.read_csv('more_recipe_urls.csv')
counter = 0
urls = []
for i in df1['recipe_urls']:
    counter += 1
    url = i
    headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36 Edg/114.0.1823.58"}
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")
    print(i)
    if soup.find_all('div', {'class': 'mntl-recipe-details__value'}) == []:
        continue
    urls.append(url)
    print(counter)
urls
```
